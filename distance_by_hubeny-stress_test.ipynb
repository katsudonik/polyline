```python
import numpy as np
from numpy.random import *
import math
import time

count = 3600 * 12
_lo = np.array(rand(count))
_la = _lo**3 

lo = np.sort(_lo + 35) 
la = np.sort(_la + 139)
points_d = np.c_[lo, la]
_points = points_d
points = points_d
points = np.delete(points, -1 , 0)
_points = np.delete(_points, 0 , 0)
v = np.c_[points, _points]
```

```python
def hubeny(latf, longf, late, longe):
    y1 = latf
    y2 = late
    x1 = longf
    x2 = longe
    
    a = 6378137.0
    b = 6356752.314140
    dy = math.radians(y2 - y1)
    dx = math.radians(x2 - x1)
    my = math.radians((y1 + y2) / 2)
    e2 = (a**2 - b**2) / a**2
    Mnum = a * (1 - e2)
    W = math.sqrt(1 - e2 * math.sin(my)**2)
    M = Mnum / W**3
    N = a / W
    d = math.sqrt((dy * M)**2 + (dx * N * math.cos(my))**2)
    print(d)
```

```python
_d = 0
for point in v:
    start = time.time()
    
    hubeny(point[0], point[1], point[2], point[3])
    elapsed_time = time.time() - start
    print ("elapsed_time:{0}".format(elapsed_time) + "[sec]")
```
